```python
print("Shape of the dataset:", df.shape)

print("\nColumns in the dataset:\n", df.columns)

print("\nMissing values in each column:\n", df.isnull().sum())

print("\nData types:\n", df.dtypes)

print("\nSummary Statistics:\n", df.describe(include='all'))
```

```python
plt.figure(figsize=(10, 6))

sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)

plt.title('Missing Values Heatmap')

plt.show()
```

```python
print("\nSmoking:\n", df['Smoking'].value_counts())

sns.countplot(x='Smoking', data=df)

plt.title('Smoking')

plt.show()




print("\nBlood Pressure:\n", df['Blood Pressure'].value_counts())

sns.countplot(x='Blood Pressure', data=df)

plt.title('Blood Pressure')

plt.show()


print("\nAge Statistics:\n", df['Age'].describe())

sns.histplot(x='Age', data=df, kde=True)

plt.title('Age Distribution')

plt.show()




print("\nAlcohol Intake:\n", df['Alcohol Intake'].value_counts())

sns.countplot(x='Alcohol Intake', data=df)

plt.title('Alcohol Intake')

plt.show()


print("\nStress Level:\n", df['Stress Level'].value_counts())

sns.countplot(x='Stress Level', data=df)

plt.title('Stress Level')

plt.show()


print("\nHeart Disease:\n", df['Heart Disease'].value_counts())

sns.countplot(x='Heart Disease', data=df)

plt.title('Heart Disease')

plt.show()
```

```python
print("\nHeart disease by Smoking:\n", df.groupby('Smoking')['Heart Disease'].value_counts())

sns.countplot(x='Smoking', hue='Heart Disease', data=df)

plt.title('Heart disease by Smoking')

plt.show()




print("\nHeart Disease by age:\n", df.groupby('Age')['Heart Disease'].value_counts())

sns.countplot(x='Age', hue='Heart Disease', data=df)

plt.title('Heart Disease by age')

plt.show()


print("\nHeart Disease by Blood Pressure:\n", df.groupby('Blood Pressure')['Heart Disease'].value_counts())

sns.countplot(x='Blood Pressure', hue='Heart Disease', data=df)

plt.title('Heart Disease by Blood Pressure')

plt.show()



print("\nHeart Disease by Alcohol Intake:\n", df.groupby('Alcohol Intake')['Heart Disease'].value_counts())

sns.countplot(x='Alcohol Intake', hue='Heart Disease', data=df)

plt.title('Heart Disease by Alcohol Intake')

plt.show()



print("\nHeart Disease by Stress Level:\n", df.groupby('Stress Level')['Heart Disease'].value_counts())

sns.countplot(x='Stress Level', hue='Heart Disease', data=df)

plt.title('Heart Disease by Stress Level')

plt.show()
```

```python
bins = [0, 12, 18, 40, 60, 80]

labels = ['Child', 'Teenager', 'Adult', 'Middle-aged', 'Senior']

df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)


plt.figure(figsize=(10, 6))

sns.countplot(x='Age_group', hue='Heart Disease', data=df, palette='magma')

plt.title('Heart Disease Based on Age Group')

plt.xlabel('Age Group')

plt.ylabel('Count')

plt.legend(title='Heart Disease', labels=['No', 'Yes'])

plt.show()
```
